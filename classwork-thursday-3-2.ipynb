### find genus, species, earliest appearance, latest appearance

```python
# open the file
infile = "/home/eeb177-student/Desktop/eeb-177/class-assignments/canids/clean_canids_data.csv"
with open(infile, 'r') as ff:
    canid_recs = ff.readlines()
```

```python
#use defaultdict
from collections import defaultdict

species_ranges = defaultdict(list)

# make a dictionary
for line in canid_recs:
    species = line.split('","')[5] #add single quotes around the double to get rid of the quotation marks
    minage = line.split('","')[10]
    maxage = line.split('","')[11]
    #print(minage)
    mean_age = (float(minage) + float(maxage))/2
    print(mean_age)
    #print(line.split(","),[5])

    species_ranges[species].append(mean_age) #add species as keys and mean ages as a new element
```

## Get the oldest and youngest ages of fossils for each species in species_ranges

```python
output = open("canid_ranges.csv", "w")

for key in species_ranges.keys():
    ages = species_ranges[key]
    minage = min(ages)
    maxage = max(ages)
    genus = key.split(" ")[0]
    species = key
    #print(genus, key, minage, maxage)
    output = "{},{},{},{}\n".format(genus, species, minage, maxage)
    print(output)
```
